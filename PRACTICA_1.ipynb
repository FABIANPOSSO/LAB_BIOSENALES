```python
import numpy as np

# a. Crear los vectores:
a = np.array([3.1, 1, -0.5, -3.2, 6])
b = np.array([1, 3, 2.2, 5.1, 1])
print("a.",a,b)

# b. Multiplicación escalar:
# ¿Qué se debe realizar para poder multiplicar los dos vectores?
dot_product = np.dot(a, b)
print(f"b. Resultado Escalar: {dot_product}","Para poder multiplicar ambos vectores, estos debian tener la misma dimensión")

# c. Multiplicación punto a punto:
element_wise = a * b
print(f"c. Resultado Punto a Punto: {element_wise}")

# d. Construir la matriz A:
A = np.array([
    [2, -1, -3],
    [4, 1.5, -2.5],
    [7.3, -0.9, 0.2]
])
print("d. A=",A)

# e. Transpuesta
AT = A.T
print("e. AT=",AT)

# f. Funciones de Numpy
# Ones: Crea matriz de unos, Round: redondea, Ceil: techo, Floor: piso.
print("f. Ejemplo ones:", np.ones((2,2)))
print("Ejemplo round:", np.round(A[2,0]))
print("Ejemplo ceil (7.3):", np.ceil(A[2,0]))
print("Ejemplo floor (7.3):", np.floor(A[2,0]))

# g. Acceso a elemento:
print("g. Elemento de Fila 1, Columna 3:", A[0, 2])

# h. Acceso a fila:
print("h. Segunda fila completa:", A[1, :])

# i. Dimensión de una matriz:
print("i. Dimensiones de A:", A.shape)

# j. Definición de primera señal
import matplotlib.pyplot as plt
n = np.arange(0, 101) # Intervalo 0 a 100
yn = np.sin(np.pi * 0.12 * n)
plt.plot(yn)
plt.title('j. Grafica 1. Primera señal')

```

```python
# k. Definición de segunda señal
import matplotlib.pyplot as plt
n = np.arange(0, 101) # Intervalo 0 a 100
yn = np.sin(np.pi * 0.12 * n)
y2n = np.cos(2 * np.pi * 0.03 * n)
plt.plot(y2n)
plt.title('k. Grafica 2. Segunda señal')

# Gráfica 3: yn y y2n
sn = yn + y2n
tn = yn * y2n
plt.figure(figsize=(10, 4))
plt.plot(n, yn, label='y[n]', color='teal')
plt.plot(n, y2n, label='y2[n]', color='orange')
plt.title('l. Grafica 3. Señales Originales')
plt.xlabel('Muestras (n)')
plt.ylabel('Amplitud')
plt.legend()
plt.grid(True)
plt.show()

# Gráfica 4: Suma y Producto
plt.figure(figsize=(10, 4))
plt.plot(n, sn, label='Suma s[n]', color='green')
plt.plot(n, tn, label='Producto t[n]', color='red')
plt.title('Grafica 4. Operaciones entre Señales')
plt.xlabel('Muestras (n)')
plt.ylabel('Amplitud')
plt.legend()
plt.grid(True)
plt.show()
```

```python
#REPASO DE PANDAS
import pandas as pd

def estadisticas_notas(notas_dict):
    """Devuelve serie con min, max, media y desviación"""
    serie = pd.Series(notas_dict)
    return pd.Series({
        'Min': serie.min(),
        'Max': serie.max(),
        'Media': serie.mean(),
        'Desv. Típica': serie.std()
    })

# Ejemplo de uso
notas_alumnos = {'Ana': 4.5, 'Luis': 3.2, 'Marta': 4.8, 'Juan': 2.5}
print(estadisticas_notas(notas_alumnos))
```

```python
import io
# Cargar datos desde el archivo adjunto
# Se usa sep=';' porque el archivo subido usa ese delimitador
df = pd.read_csv('datos.csv', sep=';')

# Mostrar primeras y últimas filas
print("Cabecera del DataFrame:\n", df.head())
print("\nFinal del DataFrame:\n", df.tail())

# Limpieza de datos
if 'Unnamed: 0' in df.columns:
    df.drop(columns=['Unnamed: 0'], inplace=True)
    print("\n--- Columna 'Unnamed: 0' eliminada exitosamente ---")
else:
    print("\n--- La columna no existe en este archivo ---")

print("\nDataFrame resultante:")
print(df.head())

# Cálculo del BMI (IMC)
# Altura / 100 para obtener metros
df['BMI'] = df['Weight'] / ((df['Height'] / 100) ** 2)
print(df.head())


# Categorización
def categorizar_imc(bmi):
    if bmi < 18.5: return 'Bajo peso'
    elif 18.5 <= bmi <= 24.9: return 'Normal'
    elif 25 <= bmi <= 29.9: return 'Sobrepeso'
    else: return 'Obesidad'

df['Categoría'] = df['BMI'].apply(categorizar_imc)

print("\nResultados del IMC:")
display(df)
```
